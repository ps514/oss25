## Data wrangling

### Subtask:
Transform the data for analysis by creating new features and potentially aggregating data.


**Reasoning**:
Create new features '총수출입금액(달러)' and '수출입비율', then group data by '국가명' and calculate the sum of '총수출입금액(달러)', finally sort the result.



```python
# Create '총수출입금액(달러)'
df['총수출입금액(달러)'] = df['당월수출입미화금액(달러)'] + df['당해누계수출입미화금액(달러)']

# Create '수출입비율'
df['수출입비율'] = df['당월수출입미화금액(달러)'] / df['총수출입금액(달러)']
df['수출입비율'].fillna(0, inplace=True) # Handle potential division by zero

# Group data and calculate sum of '총수출입금액(달러)'
df_country_totals = df.groupby('국가명')['총수출입금액(달러)'].sum().reset_index()

# Sort the result
df_country_totals = df_country_totals.sort_values(by='총수출입금액(달러)', ascending=False)

display(df_country_totals)
```
